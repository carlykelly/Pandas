```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#display(school_data_complete)
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
math_score_avg = school_data_complete["math_score"].mean()
read_score_avg = school_data_complete["reading_score"].mean()
total_budget = school_data["budget"].sum()
math_pass = school_data_complete[school_data_complete["math_score"]>=70].count()["math_score"]

#string, take out formatting if need to use number again
math_percent_string = (f"{math_pass/total_students*100:.4}%")
math_percent = (math_pass/total_students)*100
read_pass = school_data_complete[school_data_complete["reading_score"]>=70].count()["reading_score"]
combined_pass = school_data_complete[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)].count()["reading_score"] 
read_percent_string = (f"{read_pass/total_students*100:.4}%")
reading_percent = (read_pass/total_students)*100
combined_percent_string = (f"{combined_pass/total_students*100:.4}%")
combined_percent = (combined_pass/total_students)*100
```

```python
dist_summ_dict = [{"Total Schools": total_schools, "Total Students": total_students, 
                   "Total Budget": total_budget, "Average Math Score": math_score_avg, 
                  "Average Reading Score": read_score_avg, "% Passing Math": math_percent_string,
                  "% Passing Reading": read_percent_string, "% Overall Passing": combined_percent_string}]
dist_summ_df = pd.DataFrame(dist_summ_dict)
dist_summ_df["Total Budget"] = dist_summ_df["Total Budget"].map("${:,.2f}".format)
dist_summ_df["Total Students"] = dist_summ_df["Total Students"].map("{:,}".format)
dist_summ_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#reanaming exisiting columns to match final output
school_data_complete = school_data_complete.rename(columns = {"school_name":"School Name","type": "School Type", "size": "Total Students", "budget":"Total School Budget"})
```

```python
#adding Per student budget column
school_data_complete["Per Student Budget"] = school_data_complete["Total School Budget"]/school_data_complete["Total Students"]
```

```python
#finding average math & reading score per school
school_data_group = school_data_complete.groupby(["School Name", "School Type"])
school_group_avg = school_data_group.mean()

#finding total math passing per school
school_data_mathp = school_data_complete.loc[school_data_complete['math_score'] >= 70]
school_data_math_group = school_data_mathp.groupby("School Name")
num_kids_mathp = school_data_math_group.count()
#finding math pass % per school
school_group_avg["% Passing Math"] = num_kids_mathp["math_score"]/school_group_avg["Total Students"]*100
school_group_avg

#findng total reading passing per school
school_data_readp = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
school_data_read_group = school_data_readp.groupby("School Name")
num_kids_readp = school_data_read_group.count()

#finding reading pass % per school
school_group_avg["% Passing Reading"] = num_kids_readp["reading_score"]/school_group_avg["Total Students"]*100
school_group_avg

#finding total pass both per school
school_data_allp = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]
school_data_all_group = school_data_allp.groupby("School Name")
num_kids_allp = school_data_all_group.count()


#finding pass both % per school
school_group_avg["% Overall Passing"] = num_kids_allp["reading_score"]/school_group_avg["Total Students"]*100
```

```python
#dropping unneeded columns
school_group_avg = school_group_avg[["Total Students", "Total School Budget", "Per Student Budget", "math_score","reading_score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]
school_group_avg.head()
#changing column names
school_group_avg= school_group_avg.rename(columns = {"math_score": "Average Math Score", "reading_score": "Average Reading Score"})
```

```python
#Need to format
school_group_avg_form = school_group_avg.copy()
school_group_avg_form["Total Students"] = school_group_avg_form["Total Students"].astype(int)
school_group_avg_form["Total School Budget"] = school_group_avg_form["Total School Budget"].map("${:,.2f}".format)
school_group_avg_form["Per Student Budget"] = school_group_avg_form["Per Student Budget"].map("${:,.2f}".format)
school_group_avg_form["Total Students"] = school_group_avg_form["Total Students"]
school_group_avg_form.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#seperate series per grade
grade_nine = school_data_complete.loc[school_data_complete['grade'] == '9th']
grade_ten = school_data_complete.loc[school_data_complete['grade'] == '10th']
grade_eleven = school_data_complete.loc[school_data_complete['grade'] == '11th']
grade_twelve = school_data_complete.loc[school_data_complete['grade'] == '12th']
#Finding average per grade
math_nine = grade_nine.groupby("School Name").mean()['math_score']
math_ten = grade_ten.groupby("School Name").mean()['math_score']
math_eleven = grade_eleven.groupby("School Name").mean()['math_score']
math_twelve = grade_twelve.groupby("School Name").mean()['math_score']
#making each grade average a series
average_math = pd.merge(math_nine, math_ten, on = 'School Name')
average_math = average_math.rename(columns = {"math_score_x": "9th", "math_score_y": "10th"})
average_math = pd.merge(average_math, math_eleven, on = 'School Name')
average_math = pd.merge(average_math, math_twelve, on = 'School Name')
average_math = average_math.rename(columns = {"math_score_x": "11th", "math_score_y": "12th"})
average_math.index.name =None
average_math
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#making each grade average a series
read_nine = grade_nine.groupby("School Name").mean()['reading_score']
read_ten = grade_ten.groupby("School Name").mean()['reading_score']
read_eleven = grade_eleven.groupby("School Name").mean()['reading_score']
read_twelve = grade_twelve.groupby("School Name").mean()['reading_score']
average_read = pd.merge(read_nine, read_ten, on = 'School Name')
average_read = average_read.rename(columns = {"reading_score_x": "9th", "reading_score_y": "10th"})
average_read = pd.merge(average_read, read_eleven, on = 'School Name')
average_read = pd.merge(average_read, read_twelve, on = 'School Name')
average_read = average_read.rename(columns = {"reading_score_x": "11th", "reading_score_y": "12th"})
average_read.index.name = None
average_read.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 585, 630, 645, 675]
group_names = ["less than $585", "\$585-\$629", "\$630-\$644", "\$645-\$675"]
school_group_avg["Spending Ranges(Per Student)"] = pd.cut(school_group_avg["Per Student Budget"], bins, labels = group_names, 
                                                          right = False)

school_spending= school_group_avg.groupby("Spending Ranges(Per Student)")
school_spending = school_spending.mean()
school_spending = school_spending [["Average Math Score", "Average Reading Score", "% Passing Math", 
                                    "% Passing Reading", "% Overall Passing"]]
school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 1000, 2000, 5000]
group_names = ["Small(<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_group_avg["School Size"] = pd.cut(school_group_avg["Total Students"], bins, labels = group_names,
                                                include_lowest = True)

school_size= school_group_avg.groupby("School Size")
school_size = school_size.mean()
school_size = school_size[["Average Math Score", "Average Reading Score", "% Passing Math", 
                                    "% Passing Reading", "% Overall Passing"]]
school_size
```
